Add .jpg to the filename in the xml file since it is labeled wrong.

```python
import xml.etree.ElementTree as ET

def update_xml_files(directory):
    # go through every xml file in directory
    for filename in os.listdir(directory):
        if filename.endswith(".xml"):
            # full path to file
            file_path = os.path.join(directory, filename)
            tree = ET.parse(file_path)
            root = tree.getroot()

            # see if the filename ends with .jpg
            file_tag = root.find('filename')
            if file_tag is not None and not file_tag.text.endswith('.jpg'):
                file_tag.text += '.jpg'  # add .jpg to the filename

                # save XML back to file
                tree.write(file_path)
                print(f"Updated {file_path}")

update_xml_files('/content/OD-WeaponDetection/Pistol detection/xmls')
update_xml_files('/content/WeaponDetectionDataset/annotations')
```

Make .JPG endings into .jpg

```python
def rename_jpg_files(directory):
    # go through directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".JPG"):
                # old file path
                old_file_path = os.path.join(root, file)
                # new file path
                new_file_path = os.path.join(root, file[:-4] + '.jpg')

                # rename file
                shutil.move(old_file_path, new_file_path)
                print(f"Renamed '{old_file_path}' to '{new_file_path}'")

# rename directory
directory_path = '/content/OD-WeaponDetection/Weapons and similar handled objects/Sohas_weapon-Detection/images'
rename_jpg_files(directory_path)
```

Process the multiple images dataset to get rid of excess labels and rename bad labels.

```python
import xml.etree.ElementTree as ET

def remove_classes_from_xml(xml_file, classes_to_remove):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    # iterate through 'object' elements in xml and remove those that are in classes to remove
    for obj in root.findall('object'):
        class_name = obj.find('name').text
        if class_name in classes_to_remove:
            root.remove(obj)

    # save xml
    tree.write(xml_file)

classes_to_remove = ['billete', 'smartphone', 'monedero', 'tarjeta']

# directory with xml files i want to remove these classes from
xml_directory = '/content/OD-WeaponDetection/Weapons and similar handled objects/Sohas_weapon-Detection/annotations/xmls'

# go through each xml file
for xml_file in os.listdir(xml_directory):
    if xml_file.endswith('.xml'):
        remove_classes_from_xml(os.path.join(xml_directory, xml_file), classes_to_remove)
```

```python
# copy files
def copy_files(src_dir, dst_dir):
    for filename in os.listdir(src_dir):
        src_path = os.path.join(src_dir, filename)
        dst_path = os.path.join(dst_dir, filename)
        shutil.copy(src_path, dst_path)

# Copy knife images and annotations
copy_files(knife_images, combined_images)
copy_files(knife_annotations, combined_annotations)

# # Copy pistol images and annotations
copy_files(pistol_images, combined_images)
copy_files(pistol_annotations, combined_annotations)

# Copy multiple images and annotations
copy_files(multiple_images, combined_images)
copy_files(multiple_annotations, combined_annotations)

# Copy cctv images and annotations
copy_files(cctv_images, combined_images)
copy_files(cctv_annotations, combined_annotations)
```
